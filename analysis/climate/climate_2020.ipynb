```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_climate_2019 = pd.read_csv('../../data/prepared/df_climate_2019.csv')
df_climate_2021 = pd.read_csv('../../data/prepared/df_climate_2021.csv')
df_climate_2020 = pd.read_csv('../../data/prepared/df_climate_2020.csv')
```

```python
sns.lineplot(
    data=df_climate_2019,
    x='month_year',
    y='wind_speed'
)

plt.xticks(rotation=45)
```

```python
sns.lineplot(
    data=df_climate_2020,
    x='month_year',
    y='wind_speed'
)

plt.xticks(rotation=45)
```

```python
sns.lineplot(
    data=df_climate_2021,
    x='month_year',
    y='wind_speed'
)

plt.xticks(rotation=45)
```

```python
sns.lineplot(
    data=df_climate_2019,
    x='month_year',
    y='solar_radiation'
)

plt.xticks(rotation=45)
```

```python
sns.lineplot(
    data=df_climate_2020,
    x='month_year',
    y='solar_radiation'
)

plt.xticks(rotation=45)
```

```python

sns.lineplot(
    data=df_climate_2021,
    x='month_year',
    y='solar_radiation'
)

plt.xticks(rotation=45)
```

### Solar radiation
- There is a seasonality (more radiation from April to August)
- Stable range (20 - 250W/m2)

### Wind speed
- Weak seasonality to be seen (more wind in the first quarter and the last quarter)
- Range changes per year

```python
from darts import TimeSeries
import pandas as pd
import matplotlib.pyplot as plt
from darts.models import AutoARIMA
```

```python
# Read a pandas DataFrame
df = pd.read_csv('../../data/prepared/df_energy_climate_2021.csv')
```

## Exponential smoothing

```python

# Create a TimeSeries, specifying the time and value columns
series = TimeSeries.from_dataframe(df, "datetime", "energy_price")

# Set aside the 20 percent as a validation series
train, val = series[:-1800], series[-1800:]
```

```python
model_aarima = AutoARIMA()
model_aarima.fit(train)
prediction_aarima = model_aarima.predict(len(val))
```

```python
series.plot(label='actual', lw=0.5)
prediction_aarima.plot(label="forecast", lw=1)
plt.legend()
plt.xlabel('Month')
plt.ylabel('Energy price')
```
